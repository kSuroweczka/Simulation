```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mplt
import os
import cv2

path_to_folder = './data/'
file = 'density_matrix.csv'

path = os.path.join(path_to_folder, file)

with open(path, 'r'):
    df = pd.read_csv(path, sep=',', header=0, index_col=0)
    data = df.values
    data = data.transpose()
    data = np.flip(data, axis=0)

print(data.shape)
    
```

```python
# jeśli w danym punkcie w danym kroku czasowym znajduje się student warość w macierz zwiększa się o 1
# w ten sposób po zakończeniu symulacji w macierzy widać najczęściej zajmowane przez studentów miejsca

plt.figure(figsize=(20,10))

plt.axis('off')
plt.imshow(data, cmap='hot')
```
